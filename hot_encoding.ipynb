```python
# Dependancies
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
import numpy as np
```

```python
# Read in CSV file and clean the data 
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(2)
```

                                            Hot Encoding 

```python
# Reformat the data
new_data = df.values
X = new_data[:, 1:40]
y = new_data[:, 0]
print(y)
```

```python
# LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
encoded_y
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
# Convert to categorical data
from keras.utils import to_categorical

# One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

                                            Train Test Split

```python
# Splitting data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=1, stratify=y)
# startify -asking the model to take the training and test set such that the class proportion is same as of the 
# whole dataset
```

                                            Deep Learning Model

```python
# Min & Max Scaler 
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
y_train_categorical
```

```python
# Deep Learning Model 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
deep_model = Sequential()
deep_model.add(Dense(units=100, activation='relu', input_dim=40))
deep_model.add(Dense(units=100, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
deep_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
deep_model.summary()
```

```python
# Fit (train) the model
deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
# Making Predictions (The predict_classes method is only available for the Sequential class)
encoded_predictions = deep_model.predict_classes(X_test_scaled[:7])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:7])}")
```

                                        Decision Tree & Random Forest Model

```python
# Accuracy score for Decision Tree
clf = tree.DecisionTreeClassifier()

# Fittin the Decision Tree
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# Accuracy score for Random Forest Model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500)

# Fitting the training model
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Actual class predictions
rf_predictions = rf.predict(y_test)
# Probabilities for each class
rf_probs = rf.predict_proba(y_test)[:, 1]
```
